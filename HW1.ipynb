## Q3:
In Chapter $1$ example $1.4$, Grinstead and Snell talk about measuring two natural quantities associated with the so-called *simple random walk* $S_0 = 0, S_t = S_{t-1}+ X_t$, where $X_t = \pm 1$ with probability $\frac{1}{2}$ (they describe this as a simple betting game between Peter and Paul, where at each time $t$ they toss a fair coin, and if it comes up heads, then Peter wins a penny from Paul, else Paul wins a penny from Peter). We will now replicate their experiments.

**a.** Generate $n = 1000$ copies of a simple random walk $S^i = (S^i_0,S^i_1,\ldots,S^i_{T})$ (with $T=100$ and $i\in\{1,2,\ldots,n\}$). For each walk, compute $Z^i_T = \frac{S^i_T}{T}$ (i.e., the per-round winnings of Paul after $100$ games), as well as $W^i_T = \frac{1}{T}\sum_{t=1}^{T}\mathbb{1}_{\{S^i_t>0\}}$ (i.e., the fraction of rounds in $[1,T]$ where Paul is in the lead). Plot a histogram of both these quantities (using `plt.hist(data, density=True,bins='auto')`).


---

### Ans.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(17)
n = 1000
T = 100

def random_walk(T):
  x = np.random.random(T) > .5
  x = np.array(list(map(lambda x: 1 if x else -1, x)))
  s = [0]
  for i in range(1, T + 1):
    s.append(s[i - 1] + x[i - 1])
  return np.array(s)


def simulate(n):
  z = []
  w = []
  for _ in range(n):
    s = random_walk(T)
    z.append(s[-1] / T)
    w.append(np.sum(s > 0) / T)
  return {'z': z, 'w': w}

cache = simulate(n)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))
ax1.hist(cache['z'], density=True, bins='auto', color='b')
ax2.hist(cache['w'], density=True, bins='auto', color='y')
ax1.legend('Z')
ax2.legend('W')
```

**b.** Generate $10,000$ points that are uniformly distributed in a square around the origin with corners $(-1,-1),(-1,1),(1,-1)$ and $(1,1)$. To do this, independently generate the $x$- and $y$-coordinates of a point to be uniformly distributed in $(-1,1)$. Create an additional array $Z$ of length $10,000$, which indicates whether each random point lies within a circle of radius $1$ centered at the origin. Plot these points using a scatter plot with different colors for points within the circle and points outside.

*Hint: If you store your coordinate vectors as $X$ and $Y$ and the vector of indicators as $Z$, then use* <plt.scatter(X,Y, c = Z, s = 2.0, cmap=plt.cm.Paired)>


---

### Ans.
See plot below.

```python
def generate_points(n):
  """
  Generate n uniformly distributed points in with x and y in [-1, 1].
  Args:
    n: number of points
  Returns:
    x: x-coordinates of points
    y: y-cordinates of points
  """
  x = np.random.uniform(low=-1, high=1, size=n)
  y = np.random.uniform(low=-1, high=1, size=n)
  return x, y


def get_indicators(x, y):
  return np.array(list(map(lambda t: 1 if t[0]**2 + t[1]**2 <= 1 else 0, zip(x, y))))

n = 10000
x, y = generate_points(n)
z = get_indicators(x, y)

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(x, y, c=z, s=2.0, cmap=plt.cm.Paired)
```

**c.** Define $N_{circle}$ to be the (random) number of points which lie within the circle. What is $\mathbb{E}[N_{circle}]$ and $Var(N_{circle})$ (express your answer in terms of $\pi$)?

Suppose also that you did not know the value of $\pi$; can you still argue that $Var(N_{circle})\leq \frac{n}{4}$?


---

### Ans.

$N_{circle}\sim B(10000, \pi/4)$. Thus
$$
\mathbb E[N_{circle}] = 10000 \cdot \frac{\pi}{4} = 2500\pi.\\
$$
and
$$
Var(N_{circle}) = 10000 \cdot \frac{\pi}{4}\left(1 - \frac{\pi}{4}\right) = 2500\pi - 625\pi^2.
$$

Without knowing the value of $\pi$, I can still argue that $Var(N_{circle})\leq \frac{n}{4}$. To see this,
$$
\begin{align*}
Var(N_{circle}) &= \frac{\pi n}{4}\left(1 - \frac{\pi}{4}\right)\\
&= \frac{n}{4}\left(\pi - \frac{\pi^2}{4}\right).
\end{align*}
$$

Now claim that $\pi - \pi^2/4 <= 1$. To prove this, observe that
$$
\pi - \frac{\pi^2}{4} - 1 = -\left(\frac{\pi}{2} - 1\right)^2 \leq 0,
$$
thus $\pi - \pi^2/4 \leq 1$. Now
$$
Var(N_{circle}) = \frac{n}{4}\left(\pi - \frac{\pi^2}{4}\right) \leq \frac{n}{4} \cdot 1 = \frac{n}{4}
$$
Q.E.D.

**d.** Given $n$ random points generated as in part (b), we can estimate $\pi$ as $\pi_n = \frac{4N_{circle}}{\pi}$ for $\pi$ (where as before $N_{circle}$ is the random number of points that lie inside the circle). For $n$ between $1$ and $1000$, plot the relative error $|\pi_n - \pi|/\pi$. Also plot the function $1/\sqrt{n}$ on the same plot. How do the relative errors compare to this function?


---

### Ans.


```python
def get_relative_error(n):
  """
  Compute the relative error of pi when simulating with n random points.
  Args:
    n: number of random points
  Return:
    relative error
  """
  x, y = generate_points(n)
  z = get_indicators(x, y)
  N_circle = sum(z)
  pi_n = 4 * N_circle / n
  return np.abs(pi_n - np.pi) / np.pi

n = np.arange(1, 1000 + 1)
relative_error = np.array(list(map(get_relative_error, n)))
func_value = np.array(list(map(lambda x: 1 / np.sqrt(x), n)))

fig, ax = plt.subplots()
ax.plot(n, relative_error, label='relative error')
ax.plot(n, func_value, label='function value')
ax.legend()
```

Above we used simulation to compute the area of a unit circle. Using a similar process, we can find the volume of a 'sphere' in $n$ dimensions. Estimate this volume for dimensions $d\in\{2,3,\ldots,20\}$, using $100,000$ samples in each dimension. How does your estimate of the volume of the sphere behave as the dimension increases? Also, do your estimates seem smooth or noisy (and if the latter, why)?

### Ans.

```python
import numpy as np
np.random.seed(89)
def generate_ndim_points(num_points, dimension):
  coords = []
  for _ in range(dimension):
    coords.append(np.random.uniform(low=-1, high=1, size=num_points))
  return np.array(coords)

def get_ndim_indicator(coords):
  z = []
  for i in range(coords.shape[1]):
    if np.linalg.norm(coords[:, i]) ** 2 <= 1:
      z.append(1)
    else:
      z.append(0)
  return np.array(z)

def get_volume(z, dimension):
  return 2 ** dimension * sum(z) / len(z)

def compute_ndim_volume(num_points, dimension):
  coords = generate_ndim_points(num_points, dimension)
  z = get_ndim_indicator(coords)
  volume = get_volume(z, dimension)
  return volume


volume = []
for d in range(2, 20 + 1):
  volume.append(compute_ndim_volume(100000, d))

print(volume)

print("The estimated value first goes up but then goes down to zero. \nIt is noisy, because as dimension goes higher, distance of random points drawn from the sample space tends to be farther away from each other, \nso much more data points are needed to produce an accurate estimate.")
```
